```python
legal = credit[credit.Class == 0]
fraud = credit[credit.Class == 1]
```

```python
print(legal.shape)
print(fraud.shape)
```

```python
credit.groupby('Class').mean()
```

```python
legal_sample = legal.sample(n = 492)
```

```python
new_credit = pd.concat([legal_sample,fraud],axis = 0)
```

```python
new_credit.head()
```

```python
new_credit['Class'].value_counts()
```

```python
x = new_credit.drop(columns = 'Class',axis = 1)
y = new_credit['Class']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test , y_train , y_test = train_test_split(x,y,test_size = 0.2, stratify = y ,random_state = 2)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
x_train_pred = model.predict(x_train)
acc = accuracy_score(x_train_pred,y_train)
print(f'training data accuracy  = {acc}')
```

```python
x_test_pred = model.predict(x_test)
acc_test = accuracy_score(x_test_pred,y_test)
print(f'test data accuracy = {acc_test}')
```
